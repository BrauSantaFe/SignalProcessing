## **_2. Reconstruccion de la función_**

```python
import numpy as np
import matplotlib.pyplot as plt
import cmath
import math
from scipy.integrate import simps
```

<br><br> **Graficando nuestra funcion f(x):**<br>
$$
\begin{align}
f(x)=
\begin{cases}
\pi&\text{; $-\pi\leq x\leq0$}\\
x&\text{; $0\leq x\leq\pi$}\\
\end{cases}
\end{align}
$$

```python
x1=np.linspace(-np.pi,0.,100)
x2=np.linspace(0.,np.pi,100)
fx1=np.zeros(len(x1))
fx2=np.zeros(len(x2))
for i in range (len(x1)):
    fx1[i]=np.pi
for i in range (len(x2)):
    fx2[i]=x2[i]
plt.plot(x1,fx1,color="b")
plt.plot(x2,fx2,color="r")
```

_Grafica 1. Funcion f(x)_<br>
***

**A continuación se procederá a generar una gráfica continua.**

```python
muestras=1000
fx=np.zeros(muestras)
x=np.linspace(-np.pi,np.pi,muestras)
for i in range (muestras):
    if x[i]<0.:
        fx[i]=np.pi
    else: 
        fx[i]=x[i]

plt.plot(x,fx)
```

_Grafica 2. Funcion f(x) reconstruida_<br>
***

**Ahora se generará construirá la funcion f(x) utilizando diferentes armonicos en la serie de Fourier**

<br><br> **2.1 Para $ n=1,2,3,4 $ :**

```python
a0=0.75*np.pi
armc=np.zeros(muestras)
arms=np.zeros(muestras)
arm=np.zeros(muestras)
an=np.zeros(muestras)
bn=np.zeros(muestras)
w0=1
sum=0
for n in range(1,4):
    an=((-1.)**n-1)/(np.pi*n**2)
    bn=-1./n
    armc=an*np.cos(n*w0*x)
    arms=bn*np.sin(n*w0*x)
    sum=sum+(armc+arms)
    plt.plot(x,a0+sum)
plt.plot(x,a0+sum,label='n=1')
plt.plot(x,a0+sum,label='n=2')
plt.plot(x,a0+sum,label='n=3')
plt.plot(x,a0+sum,label='n=4')

plt.legend()

```

_Grafica 3. Aproximacion a f(x) por series de Fourier con 4 iteraciones_<br>
<br> Aquí podemos observar cómo nuestras iteraciones son muy similares a nuestras _Gráficas 1 y 2_.
***

<br>**2.2 Para $1\leq n\leq 100$ :**

```python
for n in range(1,100):
    an=((-1.)**n-1)/(np.pi*n**2)
    bn=-1./n
    armc=an*np.cos(n*w0*x)
    arms=bn*np.sin(n*w0*x)
    sum=sum+(armc+arms)
    plt.plot(x,a0+sum)
```

_Grafica 4. Aproximacion a f(x) por series de Fourier con 100 iteraciones._<br>
<br>En este caso a comparación de la _Gráfica 3_ se puede obervar una mayor aproximación a la función debido a la cantidad del número de muestras, es importante recalcar que tiene una mayor similitud respecto al armónico de la iteración _n=3 y n=4_.
***

<br><br>**2.3 Para $n= 1,3,5,7,...,100 $ :**

```python
for n in range(1,100,2):
    an=((-1.)**n-1)/(np.pi*n**2)
    bn=-1./n
    armc=an*np.cos(n*w0*x)
    arms=bn*np.sin(n*w0*x)
    sum=sum+(armc+arms)
    plt.plot(x,a0+sum)
```
